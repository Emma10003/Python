```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('../data/young_survey.csv')
df.head()
```

```python
interests = df.loc[:, 'History':'Pets']
interests.head()
```

```python
# 관심분야 간 상관계수
corr = interests.corr()
corr
```

```python
# 역사
corr['History'].sort_values(ascending=False)
```

```python
sns.clustermap(corr)
```
