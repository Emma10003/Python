````python
import pandas as pd
import seaborn as sns
````

````python
body_df = pd.read_csv('../data/body.csv', index_col=0)
body_df.head()
````

````python
# pandas로 산점도 그래프 그리기 - 키와 몸무게의 연관성 보기
body_df.plot(kind='scatter', x='Height', y='Weight')
````

````python
# 학생의 몸무게가 주어졌을 때, 그 학생의 키를 예측하기 - Regression Line을 그리기기
sns.lmplot(data=body_df, x='Height', y='Weight')
````
