#### 산점도 그래프
```python
* 용도: 두 변수 간 상관관계를 보여주기에 적합한 그래프.
* x와 y 모두 수치형 데이터.
1) 산점도 그래프 그리기
- df.plot(kind='scatter', x='컬럼1', y='컬럼2')
```
___

````python
%matplotlib inline
import pandas as pd
````

````python
df = pd.read_csv('../data/exam.csv')
df.head()
````

````python
# 수학 시험을 잘 보는 학생이 읽기 시험도 잘 볼까? -> 수학점수와 읽기점수에 대한 산점도 그래프
df.plot(kind='scatter', x='math score', y='reading score')
````

````python
# 수학점수와 쓰기점수
df.plot(kind='scatter', x='math score', y='writing score')
````

````python
# 읽기점수와 쓰기점수
df.plot(kind='scatter', x='reading score', y='writing score')  # 두 변수는 특히 연관성이 강하다!
````
