#### 문법 정리
```python
* 용도: 데이터의 추이를 보기 위해 사용.
* 선 그래프는 수치형 데이터에 대해서만 그릴 수 있다!
1) 기본 선 그래프 그리기
- df.plot(kind='line')
    > df.plot()의 기본형은 'line'이기 때문에 굳이 kind 파라미터를 작성하지 않아도 됨.

2) 일부 컬럼만 선택해서 그래프 그리기
- df.plot(y='컬럼명'): 컬럼 1개에 대해서만 그래프가 그려짐.
- df.plot(y=['컬럼1', '컬럼2']): 값을 리스트로 넘겨주면 여러 개의 컬럼에 대해서 그래프가 그려짐.
    > 동일한 코드:
        - df['컬럼명'].plot()
        - df[['컬럼1', '컬럼2']].plot()
```
___

````python
%matplotlib inline
import pandas as pd
````

````python
df = pd.read_csv('../data/broadcast.csv', index_col=0)
df
````

````python
# 기본 그래프 그리기
df.plot()
````

````python
# 하나의 방송사에 대해서만 그래프 그리기
df.plot(y='KBS')
````

````python
# 일부 방송사에 대해서만 그래프 그리기
df.plot(y=['KBS', 'JTBC'])
````

````python
# 위와 같은 코드
df['KBS'].plot()  # df['KBS']는 Series. Pandas Series형도 plot을 그릴 수 O
df[['KBS', 'JTBC']].plot()  # df[['KBS', 'JTBC']]는 DataFrame.
````
