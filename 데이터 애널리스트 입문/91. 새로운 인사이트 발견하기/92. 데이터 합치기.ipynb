#### 데이터 합치기
```python
: merge
기본 형식: pd.merge(df1, df2, on='기준컬럼')
  > 기본은 inner join

1) inner join
: 교집합만 합친다. -> 기준 컬럼의 데이터 중 겹치는 데이터만 남김.
- pd.merge(df1, df2, on='기준컬럼')

2) left outer join
: 왼쪽 데이터프레임에 존재하는 것만 합친다. -> 왼쪽 데이터프레임의 데이터는 모두 보존. (빈 곳은 NaN: Not a Number) 
- pd.merge(df1, df2, on='기준컬럼', how='left')

3) right outer join
: 오른쪽 데이터프레임에 존재하는 것만 합친다. -> 오른쪽 데이터프레임의 데이터는 모두 보존. (빈 곳은 NaN)
- pd.merge(df1, df2, on='기준컬럼', how='right')

4) full outer join
: 합집합만 합친다. -> 모든 데이터프레임의 데이터가 보존된다. (빈 곳은 NaN)
- pd.merge(df1, df2, on='기준컬럼', how='outer')
```
___

````python
import pandas as pd

price_df = pd.read_csv('../data/vegetable_price.csv')
quantity_df = pd.read_csv('../data/vegetable_quantity.csv')
````

````python
price_df
````

````python
quantity_df
````

````python
pd.merge(price_df, quantity_df, on='Product')  # inner join
````

````python
pd.merge(price_df, quantity_df, on='Product', how='left')  # left join
````

````python
pd.merge(price_df, quantity_df, on='Product', how='right')  # right join
````

````python
pd.merge(price_df, quantity_df, on='Product', how='outer')  # full outer join
````
