#### 문자열 필터링
```python
1) 특정 문자열을 포함하는 행 추출
: .str.contains('문자열')
- df.['컬럼명'].str.contains('문자열') -> 불린값으로 이뤄진 Pandas Series.
- df[df.['컬럼명'].str.contains('문자열')] -> 해당 문자열을 포함하는 행만 존재하는 데이터프레임.

2) 특정 문자열로 시작하는 행 추출
: .str.startswith('문자열')
- df[df.['컬럼명'].str.startswith('문자열')] -> 해당 문자열로 시작하는 행만 존재하는 데이터프레임.
```
___

````python
import pandas as pd

df = pd.read_csv('../data/albums.csv', encoding='latin1')
df.head(20)
````

````python
df['Genre'].unique()
````

#### Blues 장르들만 뽑기

````python
df[df['Genre'] == 'Blues']  # 장르가 'Blues'인 것만 인덱싱 됨
````

````python
df[df['Genre'].str.contains('Blues')].head()  # 장르에 Blues가 포함되어 있는 장르가 모두 나옴.
````

#### Blues가 앞에 있는 장르들만 추출

````python
df[df['Genre'].str.startswith('Blues')].head()
````

#### 새로운 행 만들기 - Blues를 포함하면 True, 아니면 False

````python
df['Contains Bluse'] = df['Genre'].str.contains('Blues')
df.head()
````
