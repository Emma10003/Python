### 직업 탐구하기 II

```python
import pandas as pd

df = pd.read_csv('../data/occupations.csv')
df.head()
```

여성 비율이 높은 순으로 직업 나열. Series 형태로!

```python
total_counts = df['occupation'].value_counts()
f_counts = df.loc[(df['gender']=='F'), 'occupation'].value_counts()

f_ratio = f_counts / total_counts
f_ratio.sort_values(ascending=False)
```

다른 방법 (groupby 사용)

```python
occ_groups = df.groupby('occupation')
# groupby의 mean()은 숫자형 데이터만 가능하므로 'gender' 열의 데이터를 0과 1로 바꿔줌.
df.loc[df['gender']=='M', 'gender'] = 0
df.loc[df['gender']=='F', 'gender'] = 1

df
```

```python
occ_groups.mean()  # 남성은 0, 여성은 1로 설정했으므로 
                   # 평균이 0에 가까울수록 남성의 비율이 높고, 1에 가까울수록 여성의 비율이 높음.
                   # 평균을 사용한 것은, 평균 = (1의 개수) / (전체의 개수)가 되기 때문에,
                   # 평균값이 곧 여성의 비율이 됨.
```

```python
occ_groups.mean()['gender'].sort_values(ascending=False)
```
