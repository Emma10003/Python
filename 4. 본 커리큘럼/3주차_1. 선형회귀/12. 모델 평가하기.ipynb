우리는 계속해서 세타 값을 조율하면서 더 좋은 가설 함수를 만들려고 하잖아요?  
그리고 결국에는 데이터에 가장 잘 맞는 최적선을 찾으려고 하는 건데요.

이런 가설 함수는 세상에 일어나는 상황을 수학적으로 표현한다는 의미에서 '**모델**'이라고 부릅니다.  
원래 마네킹이나 모형처럼 무언가를 표현하려고 하는 걸 영어로는 모델이라고 하거든요.

앞으로 머신 러닝을 하다 보면 '모델'이라는 용어를 자주 접하게 되실 거고요, 데이터를 이용해서 모델을 개선시키는 걸  
'모델을 학습시킨다'라고 할 겁니다.

## 모델 평가하기

만든 모델이 얼마나 예측을 잘 하는 모델인지를 알기 위해, 모델을 평가해야 한다.  
주로 **RMSE**(평균 제곱근 오차)를 지표로 사용한다.  

$$
\sqrt[\text{}]{MSE(평균제곱오차)}
$$

MSE만 사용하면 단위가 제곱이 되기 때문에, 원래의 단위로 사용하기 위해 제곱근을 해주는 것.

여기서, 학습 데이터로 평가를 하는 경우 당연히 RMSE는 낮게 나올 수 밖에 없다.  
(이미 해당 데이터를 바탕으로 최적화된 상태이기 때문)  

따라서, <span style='color:yellowgreen'>학습과 평가를 위한 데이터를 나누는</span> 게 일반적.  

- 학습 데이터 (training set) : 모델을 학습시킬 때 사용하는 데이터.
- 평가 데이터 (test set): 모델의 성능을 평가할 때 사용하는 데이터.  

학습에 사용된 데이터와 평가에 사용되는 데이터가 다르므로 더 신빙성 있는 평가가 가능하다.
