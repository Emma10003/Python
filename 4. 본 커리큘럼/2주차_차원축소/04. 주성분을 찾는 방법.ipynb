PCA는 고차원 데이터의 특성을 최대한 잘 반영하는 축인 주성분을 찾아서 데이터를 저차원으로 투영하는 방법이라고 했습니다.  
이때, 주성분은 어떻게 찾을 수 있을까요? 이전 레슨을 통해 데이터들의 중심(원점)을 지나면서  
모든 데이터들에서의 수직 거리의 합이 가장 가깝도록 하는 선을 찾으면 된다고 간단하게 언급해 드리긴 했는데요.  
이번 레슨을 통해 조금 더 보충해서 설명해 드리도록 하겠습니다.

먼저, 표준화를 통해 변수 간 단위를 맞추고, 원점 이동까지 한 데이터가 아래와 같이 있다고 해볼게요.

![image](image/04-이미지1.png)

주어진 데이터의 특징을 나타내면서 원점을 지나는 선은 정말 많습니다. 어떠한 선도 데이터의 특성을 일부 설명해 줄 수 있는 것이죠.  
하지만, 그 중에서도 가장 데이터의 특성을 잘 나타내는 선은 하나입니다. 그리고, 그 하나의 선을 찾는 기준은  
<span style='color:yellowgreen'>모든 데이터와 찾아진 선 사이의 거리 합이 가장 작아지도록</span> 하는 것입니다.

예를 들어서, 아래의 선도 데이터가 가진 정보 중 일부를 표현해 주고 있는데요.

![image](image/04-이미지2.png)

하지만, 데이터와 선 사이의 거리가 최소는 아닌 거 같습니다.  
위의 이미지보다는 아래 이미지에서 데이터와 선 사이의 거리의 합이 조금 더 작은 거 같네요.

![image](image/04-이미지3.png)

한 눈에 보기에도 각 데이터에서 그려진 축에 직교하는 거리의 합이 이전 이미지보다 훨씬 작죠?  
또, 직관적으로 보기에도 그린 선이 데이터가 가진 정보를 조금 더 많이 반영하고 있는 것처럼 보입니다.  
원래 데이터가 분포한 형태와 가장 비슷하게 퍼져있도록 데이터가 투영되었으니까요.  
이렇게, <span style='color:yellowgreen'>데이터와 성분 사이의 수직 거리가 더 가까울수록 해당 축이 데이터의 분포를 잘 반영</span>하게 됩니다.

그리고, 데이터와 주성분 사이의 거리가 최소가 되도록 선을 찾으면 주성분으로 데이터들을 투영했을 때 해당 데이터가 퍼져있는 정도,  
즉 <span style='color:yellowgreen'>분산이 커집니다</span>.  
그 이유는 피타고라스의 정의를 기준으로 생각하면 쉽게 이해할 수 있는데요. 아래 이미지를 참고해 보면  
데이터와 원점 사이의 거리인 C는 고정이기 때문에, 삼각형의 높이 부분에 해당하는 데이터와 선 사이의 거리(B)가 작아지면  
선의 길이, 즉 분산(A)이 늘어나는 형태인 것이죠.

![image](image/04-이미지4.png)

즉, PCA에서 데이터의 특징을 가장 잘 설명하는 주성분을 찾는 것은 데이터들과 선 사이의 수직 거리 합이 최소가 되는 선을  
찾으면 되는 것이고, 그것은 곧 임의의 선에 데이터를 투영했을 때 데이터들의 분산이 최대가 되는 선을 찾는 것과  
같은 문제가 되는 것입니다.  
<span style='color:yellowgreen'>데이터들의 분산이 최대가 되도록 하는 선이 데이터에 대한 설명력이 가장 높은 선</span>이 되는 것이죠.  
실제로 PCA연산에서는 투영하였을 때 데이터들 사이의 분산이 최대가 되도록 하는 선을 찾는 방식으로 주성분을 찾습니다.

다음으로, 두 번째 주성분은 첫 번째로 찾은 메인 주성분에 수직이면서 원점을 지나고, 데이터들과의 거리 합이 가장 작은 선을 찾으면 됩니다.

![image](image/04-이미지5.png)

주성분끼리 수직으로 직교하게 하는 이유는 <span style='color:yellowgreen'>새롭게 찾아지는 두 축이 서로 상관이 없도록 하여  
조금 더 많은 정보를 살린 상태로 차원을 축소하기 위함</span>입니다.  
예를 들어, 아래와 같이 두 번째 주성분이 그려진다면 데이터에 대하여 첫 번째 주성분과 비슷한 정보를 더 많이 나타내게 되겠죠?

![image](image/04-이미지6.png)

차원이 높아지더라도 지금 배운 내용이 그대로 적용됩니다.  
예를 들어, 3차원인 데이터를 PCA를 이용하여 차원 축소하는 상황을 가정해 볼게요.  
그럴 경우, 3차원 데이터의 특성을 가장 잘 반영하는 주성분 PC1, PC2, PC3를 위에서 설명한 방법대로  
차례대로 찾아가면 됩니다.  
그리고, 1차원으로 차원을 줄이고 싶다면 PC1으로만 데이터를 투영하면 되고, 2차원으로 차원을 줄이고 싶다면  
PC1과 PC2까지 2개의 축으로 데이터를 투영하면 되는 것이죠.
