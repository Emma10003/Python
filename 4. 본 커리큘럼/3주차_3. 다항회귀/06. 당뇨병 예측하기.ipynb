## I. 문제 만들기

```python
# 필요한 라이브러리 import
from sklearn import datasets
from sklearn.preprocessing import PolynomialFeatures

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

import pandas as pd  
```

```python
diabetes_dataset = datasets.load_diabetes()  # 데이터 셋 갖고오기

# 다항회귀를 위해 가상의 속성 만들기
polynomial_transformer = PolynomialFeatures(2)

polynomial_data = polynomial_transformer.fit_transform(diabetes_dataset.data)  # 파라미터로 데이터를 넘겨줌.

# 속성 이름 저장
polynomial_feature_names = polynomial_transformer.get_feature_names_out(diabetes_dataset.feature_names)
```

```python
X = pd.DataFrame(polynomial_data, columns = polynomial_feature_names)
# 테스트 코드
X.head()
```

```python
# 목표 변수
y = pd.DataFrame(diabetes_dataset.target, columns=['diabetes'])

X = pd.DataFrame(polynomial_data, columns = polynomial_feature_names)
```

## 모델 학습하기

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)

model = LinearRegression()  # 모델 생성
model.fit(X_train, y_train)  # 모델 학습

y_test_predict = model.predict(X_test)  # 예측
```

```python
# 모델 평가 - RMSE
mse = mean_squared_error(y_test, y_test_predict)
mse ** 0.5
```
