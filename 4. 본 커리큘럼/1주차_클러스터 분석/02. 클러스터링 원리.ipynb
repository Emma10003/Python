### 클러스터링의 원리
- 유사한 데이터는 같은 클러스터로 묶는다.  
- 유사하지 않은 데이터는 다른 클러스터로 묶는다.  

이 때, '유사한 데이터'라는 것은 어떻게 정의할 수 있을까요? 이번 레슨에서는 직관적인 설명을 위해  
'거리가 가까운 데이터가 유사하다'라는 기준을 사용해 볼게요. 해당 기준으로 클러스터링의 원리를 다시 표현하면 아래와 같습니다.

- 거리가 가까운 데이터는 같은 클러스터로 묶는다.  
- 거리가 먼 데이터는 다른 클러스터로 묶는다.

클러스터링을 할 때 **어떤 기준을 사용하냐**에 따라 결과가 다르게 나오는데요, 이번 토픽에서 그런 다양한 기준들을 하나씩 살펴보겠습니다.

먼저, 챕터 2에서는 **거리를 기준으로 클러스터를 나누는 방법**에 대해 살펴볼게요.  
거리 기반 클러스터링 모델 방법인 k-means를 사용해 보면서 컴퓨터가 어떻게 가까이에 있는 데이터들을 나누는지 알아보겠습니다.

좀 더 다양한 클러스터링 기준들은 챕터 3에서 살펴볼게요. 각 방법들에 어떤 컨셉이 적용되는지를 간략하게 살펴보고, 데이터를 통해 직접 사용해 보겠습니다.
