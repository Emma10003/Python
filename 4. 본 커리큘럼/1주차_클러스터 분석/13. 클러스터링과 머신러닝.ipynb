## 클러스터링과 머신러닝
지금까지 여러 클러스터링 모델들에 대해 배워 봤는데요. 사실, 클러스터링은 머신러닝 기법 중 비지도 학습(Unsupervised Learning)에 해당합니다.

머신러닝은 컴퓨터가 데이터를 통해 스스로 규칙을 학습하도록 하는 인공지능의 한 방법인데요.  
그 중에서도 비지도 학습은 정답(레이블)을 주지 않은 상태에서 데이터의 특성만 가지고 스스로 규칙을 찾아내는 방식입니다.

그동안 배운 모델들도 각 데이터가 어떤 클러스터에 포함되는지를 미리 지정해 주지 않고 주어진 데이터의 특성만으로 구분했었죠?  
그래서 클러스터링도 비지도 학습에 포함되는 것입니다.

## 클러스터링과 분류(Classification)
클러스터링과 비슷하게 데이터들의 특성을 바탕으로 특정 범주로 구분해 주는 분류 문제도 있습니다.

예를 들어서, 100명의 학생이 공부한 시간과 시험에 통과했는지 여부를 체크한 데이터가 있다고 해 볼게요.  
이 데이터를 바탕으로, 새로운 학생이 몇 시간 공부했는지에 따라 시험에 통과할 수 있을지를 예측해 볼 수 있겠죠?

해당 상황에서는 구분하고 싶은 정답(시험 통과 여부)가 데이터에 이미 존재합니다. 그리고, 변수(공부 시간)과 정답 사이의 규칙성을  
모델이 미리 학습하여 새로운 데이터를 합격과 불합격 범주 중 하나로 분류하는 것인데요.

데이터에 미리 정답 레이블이 있는 상태로 모델을 학습시키기 때문에, 이 경우는 지도 학습(Supervised Learning)에 해당됩니다.

클러스터링과 분류는 <span style='color:yellowgreen'>데이터의 특성에 따라 범주를 구분</span>한다는 점이 유사합니다.  

하지만, 클러스터링은 정답이 없는 상황에서 <span style='color:palegoldenrod'>순수하게 데이터의 특성만으로 구분</span>하고,  
분류는 데이터의 특성과 정답 사이의 관계를 학습하여 <span style='color:yellowgreen'>새로운 데이터가 어떤 범주에 해당할지 구분</span>합니다.

이러한 방식의 차이 때문에 둘의 **결과**에도 차이가 있는데요. <span style='color:yellowgreen'>분류 모델의 결과는 특정한 범주에 할당</span>됩니다.  
위의 예시에서도, 공부 시간에 따른 결과가 ‘합격’ 또는 ‘불합격’으로 정해지겠죠?

하지만 클러스터링은 <span style='color:palegoldenrod'>데이터들이 특정 클러스터에 할당</span>될 뿐, 그 클러스터가 무엇을 의미하는지가 함께 결과로 나오진 않습니다.  
해당 클러스터가 무엇을 의미하는지는 분석하는 사람이 별도로 해석해야 하죠.

이렇게, 데이터의 특성에 따라 적용하는 방법이 다르기 때문에 어떤 상황에서 어떤 방법론을 적용해야 할지 잘 판단해야 합니다.  

해당 내용은 머신러닝과 분류 모델에 대하여 학습하면 조금 더 명확하게 이해가 될텐데요.  
이어서 머신러닝 관련 토픽들을 수강하시면서 클러스터링과 분류를 적절한 상황에  
사용할 수 있게 학습해 보시기 바랍니다.
