#### 유일성
```python
1) row 중복
- df.index.value_counts()
    > 중복되는 인덱스 찾기
- df.drop_duplicate(inplace=True/False)
    > 중복되는 row 제거
    > 기본 설정: inplace = False

2) column 중복
- df.T
    > row와 column의 위치를 변경.
- df.T.drop_duplicate()
    > row와 column의 위치가 바뀐 상태에서 중복되는 row(원래 column)를 삭제.
    > inplace=True 쓸 수 없음.
- df.T.drop_duplicate().T
    > 중복 삭제 후 다시 원상태로 돌리기
    > 변경사항이 원본에 저장되지 않고, inplace 파라미터도 쓸 수 없으므로, 꼭 데이터프레임에 다시 저장해 줘야 함.
    > df = df.T.drop_duplicate().T
```

````python
import pandas as pd
````

````python
df = pd.read_csv('../data/dust.csv', index_col=0)
df.head()
````

#### row 중복

````python
df.index
````

````python
df.index.value_counts()
````

````python
df.loc['07월 31일']  # 중복된 row -> 삭제
````

````python
# 중복 row 삭제하기
df.drop_duplicates().head()
````

#### column 중복

````python
df.T  # row와 column이 뒤바뀜
````

````python
df.T.drop_duplicates()  # row와 column이 바뀐 상태에서 중복되는 row(원래 column)을 삭제.
````

````python
df = df.T.drop_duplicates().T  # 다시 원본처럼 row와 column의 위치가 바뀜.
````
