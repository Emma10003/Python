### 카드사 고객 분석

데이터의 중요성을 깨달은 "삼송카드"와 "현디카드"가 협업을 하기로 결정했습니다.

두 카드사는 사람들이 요일별로 지출하는 평균 금액을 "요일", "식비", "교통비", "문화생활비", "기타" 카테고리로 정리해서
우리에게 공유해 주기로 했는데요. 각각 `samsong.csv`파일과 `hyundee.csv`파일을 보냈습니다.

두 회사의 데이터를 활용해서, 사람들의 요일별 문화생활비를 분석해보려 합니다.

```python
%config Completer.use_jedi = False
```

```python
import pandas as pd

samsong_df = pd.read_csv('../data/samsong.csv')
hyundee_df = pd.read_csv('../data/hyundee.csv')
```

#### 내 풀이

```python
# samsong_df와 hyundee_df 병합
merge_df = pd.merge(samsong_df, hyundee_df, how='left', on='요일')   # how 파라미터 - join 방법 / on 파라미터 - 병합 기준 컬럼
merge_df
```

```python
# 병합한 데이터프레임에서 '요일', '문화생홟비' 컬럼만 보기
culture_df = merge_df.loc[:, ('요일', '문화생활비_x', '문화생활비_y')]
culture_df
```

```python
# 열 이름 변경하기
result_df = culture_df.rename(columns={'요일':'day', '문화생활비_x':'samsong', '문화생활비_y':'hyundee'})
result_df
```

#### 모범답안(인덱싱 사용)

```python
# 인덱싱 사용 - 각 데이터셋에서 필요한 컬럼 가져오기
samsong_df['요일']
samsong_df['문화생활비']
hyundee_df['문화생활비']
```

```python
# 파이썬 딕셔너리 만들기
combined_df = pd.DataFrame({
    'day': samsong_df['요일'],
    'samsong': samsong_df['문화생활비'],
    'hyundee': hyundee_df['문화생활비']
})
combined_df
```
