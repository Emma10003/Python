#### 카테고리로 분류
```python
: .map(변수명)

- df['컬럼명'].map(변수명)
    > 사전에 컬럼 내 데이터들과 대응되는 값이 딕셔너리에 값으로 정의되어 있어야 하고,
      그 사전을 '변수명'에 담아서 .map()의 파라미터로 넘겨줘야 함.
    > (key-컬럼 데이터, value-바꿀 데이터)
    > 결과: 해당 컬럼의 원래 데이터들이 딕셔너리에서 대응되는 값으로 변경된 Pandas Series가 반환됨.
    > 원래의 데이터프레임을 변경하지는 않음.

- df['새로운 컬럼'] = df['컬럼명'].map(변수명)
    > 새로운 컬럼에 Series들이 추가됨.
```
