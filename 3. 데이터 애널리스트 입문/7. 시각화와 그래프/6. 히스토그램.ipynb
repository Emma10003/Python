___
#### 히스토그램
```python
* 용도: 연속형 데이터를 범위를 정해서 그려야 할 때, 데이터 분포를 볼 때 사용.
* y는 수치데이터.
1) 히스토그램 그리기
- df.plot(kind='hist', y='컬럼명') : 기본 10개의 막대가 생성됨.

2) 히스토그램 막대 개수 설정
- df.plot(kind='hist', y='컬럼명', bins=15) : 15개의 막대가 생성됨.
    > bins 파라미터로 개수 넘겨주기.

```
___

````python
%matplotlib inline
import pandas as pd
````

````python
df = pd.read_csv('../data/body.csv', index_col=0)
df.head(10)
````

````python
# 키에 대한 히스토그림 그리기
df.plot(kind='hist', y='Height')
````

````python
# 막대 개수 정하기
df.plot(kind='hist', y='Height', bins=15)
````
