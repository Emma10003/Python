___
#### 막대 그래프
```python
* 용도: 카테고리를 특정 기준으로 비교하기 위해 사용.
1) 막대 그래프 그리기
- df.plot(kind='bar')
- df.plot(kind='barh') : 가로로 누운 막대그래프가 그려짐 (horizontal)

2) 두 데이터를 쌓아서 보기
- df.plot(kind='bar', stacked=True)

3) 특정 컬럼의 데이터만 막대 그래프로 그리기
- df['컬럼명'].plot(kind='bar')
```
___

````python
%matplotlib inline
import pandas as pd
````

````python
df = pd.read_csv('../data/sports.csv', index_col=0)
df
````

````python
df.plot()  # 현재 데이터에는 선 그래프가 적합하지 X!
````

````python
# 막대 그래프 그리기
df.plot(kind='bar')
````

````python
# 가로로 누운막대 그래프 그리기
df.plot(kind='barh')
````

````python
# 두 데이터가 위로 쌓인 막대 그래프 그리기기
df.plot(kind='bar', stacked=True)
````

````python
# 여성에 대한 데이터만 그래프로 그리기
df['Female']  # Pandas Series
df['Female'].plot(kind='bar')
````
